```python
import pandas as pd
import numpy as np

df = pd.DataFrame([
    ['blue', 'SM', 15.3, 'class2'],
    ['green', 'M', 10.1, 'class1'],
    ['red', 'L', 13.5, 'class1'],
    ['yellow', 'XL', 12.3, 'class2'],
    ['cyan', 'XXL', 15.3, 'class2'],
    ['blue', 'XXXL', 11.3, 'class2'],
    ['blue', 'XXXL', 11.3, 'class2'],
    ['blue', 'XXXL', 11.3, 'class2'],
])

df.columns = ['color', 'size', 'price', 'class_label']
df
```

```python
size_mapping = {}

for idx, size in enumerate(np.unique(df['size'])):
    size_mapping[size] = idx

print(size_mapping)
```

```python
size_mapping = {size: idx for idx, size in enumerate(np.unique(df['size']))}
size_mapping
```

```python
df['size'] = df['size'].map(size_mapping)
df
```

```python
ohe = pd.get_dummies(df['color'])
df = df.drop(columns=['color'])
```
