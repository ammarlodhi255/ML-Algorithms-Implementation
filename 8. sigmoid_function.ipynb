```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))
```

```python
z = np.arange(-10, 10, 0.1)
phi_z = sigmoid(z)
plt.plot(z, phi_z)
plt.axvline(0.0, color='k') # Puts a vertical line at x = 0.0
plt.xlabel('z')
plt.ylabel('$\phi (z)$')
plt.yticks([0, 0.5, 1])
ax = plt.gca()
ax.yaxis.grid(True)
plt.tight_layout()
plt.show()
```
