```python
df = pd.read_csv('/content/drive/MyDrive/Datasets/covtype.data', header=None)
pd.set_option('display.max_columns', len(df.columns))
df.head(5)
```

```python
cover_types = [
    'Spruce/Fir',
    'Lodgepole Pine',
    'Ponderosa Pine',
    'Cottonwood',
    'Aspen',
    'Douglas-fir',
    'Krummholz',
]

col_names = [
    'Elevation Quantitative',
    'Aspect Azimuth Aspect',
    'Slope Degrees',
    'Horizontal_Distance_To_Hydrology',
    'Vertical_Distance_To_Hydrology',
    'Horizontal_Distance_To_Roadways',
    'Hillshade_9am',
    'Hillshade_Noon',
    'Hillshade_3pm',
    'Horizontal_Distance_To_Fire_Points',
    'Rawah Wilderness Area',
    'Neota Wilderness Area',
    'Comanche Peak Wilderness Area',
    'Cache la Poudre Wilderness Area',
]

soil_types = ['Soil Type {}'.format(i) for i in range(1, 41)]
col_names.extend(soil_types)
col_names.append('Cover Type')
```

```python
df.columns = col_names
df.head(5)
```

```python
df.shape
```

```python
current_instances = df.shape[0]
print(f'Number of current instances: {current_instances}')
```

```python
# Displaying if there are any missing values 

(df == None).sum()
```

```python
df.isnull().sum()
```

```python
# If there were any missing values in the categorical column i.e. cover type column
# Following is how we would drop the rows which had missing values for the categorical column cover type

df.dropna(subset=['Cover Type'])
df.shape[0]
```

```python
# If we had the missing values for numeric columns then this is how we would handle those:
df.fillna(df.mean())
```

```python
print(f'Number of instances before and after:\nBefore: {current_instances}\nAfter: {df.shape[0]}')
```

```python
X = df.drop(columns=['Cover Type']).values
y = df['Cover Type'].values
```

```python
inv_target_mapping = {(idx + 1): target for idx, target in enumerate(cover_types)}
df['Cover Type'] = df['Cover Type'].map(inv_target_mapping)
print('Class Distribution:')
df['Cover Type'].value_counts()
```

```python
# Bringing features on the same scale through normalization using min-max scalar
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
X = mms.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, stratify=y)
```

```python
from sklearn.linear_model import SGDClassifier
import warnings
warnings.filterwarnings('ignore')

learning_rates = [0.0001, 0.001, 0.01, 0.1]
sgd_accuracies = []

for lr in learning_rates:
    sgd = SGDClassifier(eta0=lr, max_iter=50)
    sgd.fit(X_train, y_train)
    accuracy = round(sgd.score(X_test, y_test), 2)
    sgd_accuracies.append(accuracy)
    
sgd_accuracies
```

```python
def find_best_acc(accuracies):
    acc_index = 0

    for index, acc in enumerate(accuracies):
        if acc > accuracies[acc_index]:
            acc_index = index
    
    return acc_index
```

```python
max_index = find_best_acc(sgd_accuracies)
sgd_accuracy = sgd_accuracies[max_index]
print(f'SGD Accuracy: {sgd_accuracy} with Learning rate = {learning_rates[max_index]}')
```

```python
from sklearn.linear_model import Perceptron

perceptron_accuracies = []

for lr in learning_rates:
    perceptron = Perceptron(eta0=lr, max_iter=50)
    perceptron.fit(X_train, y_train)
    accuracy = round(perceptron.score(X_test, y_test), 2)
    perceptron_accuracies.append(accuracy)

perceptron_accuracies
```

```python
max_index = find_best_acc(perceptron_accuracies)
ppn_accuracy = perceptron_accuracies[max_index]
print(f'Perceptron Accuracy: {ppn_accuracy} with Learning rate = {learning_rates[max_index]}')
```

```python
from sklearn.svm import SVC

svclassifier = SVC(kernel='linear', max_iter=100)
svclassifier.fit(X_train, y_train)
svm_accuracy = svclassifier.score(X_test, y_test)
print(f'SVM Accuracy: {svm_accuracy}')
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
dt_accuracy = round(dt.score(X_test, y_test), 2)
print(f'Decision Tree Accuracy: {dt_accuracy}')
```

```python
from sklearn.linear_model import LogisticRegression

# Using solver='saga' since it is recommended to use this solver for larger datasets
logreg = LogisticRegression(solver='saga')
logreg.fit(X_train, y_train)
log_reg_accuracy = round(logreg.score(X_test, y_test), 2)
print(f'Logistic Regression Accuracy: {log_reg_accuracy}')
```

```python
from sklearn.neighbors import KNeighborsClassifier
neighbors = [3, 5, 7, 9, 11] 
knn_accuracies = []

for neighbor in neighbors:
    knn = KNeighborsClassifier(n_neighbors=neighbor)
    knn.fit(X_train, y_train)
    print(neighbor)
    # I had to test KNN on smaller number of test samples, since KNN was taking a huge amount of time
    accuracy = round(knn.score(X_test[:5000], y_test[:5000]), 2)
    knn_accuracies.append(accuracy)

knn_accuracies
```

```python
import matplotlib.pyplot as plt

# Let us plot the number of neighbors and the corresponding accuracy we achieved

plt.plot(neighbors, knn_accuracies, marker='o')
plt.xlabel('n_neighbors')
plt.ylabel('Accuracy')
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
knn_accuracy = knn_accuracies[0]
```

```python
# Plotting the accuracies of all models

x = list(range(1, 7))
y = [sgd_accuracy, ppn_accuracy, svm_accuracy, log_reg_accuracy, dt_accuracy, knn_accuracy]
my_ticks = ['SGD', 'PPN', 'SVM', 'Logreg', 'Decision Tree', 'KNN']
plt.plot(x, y, marker='o')
plt.xticks(x, my_ticks)
plt.xlabel('Classifier')
plt.ylabel('Accuracy')
plt.tight_layout()
plt.grid(True)
plt.show()
```
