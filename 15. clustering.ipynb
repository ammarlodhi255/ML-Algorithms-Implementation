```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

X, y = make_blobs (
    n_samples=180,
    n_features=2,
    centers=2,
    cluster_std=0.5,
    shuffle=True,
    random_state=42
)


plt.scatter(
    X[:, 0],
    X[:, 1],
    c='green',
    marker='o',
    edgecolor='black',
    s=50
)

plt.grid()
plt.tight_layout()
plt.show()
```
