```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

X, y = make_blobs (
    n_samples=180,
    n_features=2,
    centers=2,
    cluster_std=0.5,
    shuffle=True,
    random_state=42
)


plt.scatter(
    X[:, 0],
    X[:, 1],
    c='green',
    marker='o',
    edgecolor='black',
    s=50
)

plt.grid()
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans (
    n_clusters=2, # Specify the number of clusters.
    init='random', # Randomly pick the centroids.
    n_init=10, # Run the algorithm for 10 times independently with random centroids, and pick the one with least SSE.
    max_iter=200, # 200 iterations for each run.
    tol=1e-04, # Tolerance for SSE.
    random_state=42
)

y_kmeans = kmeans.fit_predict(X)
```
